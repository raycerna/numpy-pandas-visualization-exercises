```python
#import pyplot module as alias plt
import matplotlib.pyplot as plt  

#import other libraries 
import numpy as np
import math
from random import randint
import random
import pandas as pd
```

```python
#create range for x
x = range(-100, 100)
#copy equation for y over
y = [(x**2) - (x + 2) for x in x]

#plot x and y
plt.plot(x,y)

#adding title of equation
plt.title('y=x^2-x+2')

#plotting x and y
plt.xlabel('x')
plt.ylabel('y')

#adding origin set to 0
plt.text(0,0,'origin', color = 'r', fontsize = 25)
```

```python
#create range for x
x = range(0,100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)


plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');
```

```python
x = range(0,20)
y = [n**3 for n in x]

plt.plot(x,y)


plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
x = range(0,15)
y = [2 ** n for n in x]

plt.plot(x,y)


plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--');
```

```python
#specify range of x values
x = range(0,15)


y = [1/(n+1) for n in x]


plt.figure(figsize = (10,6))

# plot the data
plt.plot(x,y)


# Customize the plot
# Add title
plt.title('$ y = 1/(x+1)$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')


plt.grid(True, ls = '--')

plt.show()
```

```python
plt.figure(figsize = (8,5))

plt.subplot(221)
x = range(0,100)
y = [math.sqrt(n) for n in x]

plt.plot(x,y)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)
plt.grid(True, ls = '--')


plt.subplot(222)

x = range(0,20)
y = [n**3 for n in x]

plt.plot(x,y)


plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')

plt.subplot(223)
x = range(0,15)
y = [2 ** n for n in x]
plt.grid(True, ls = '--')
plt.plot(x,y)


plt.subplot(224)
x = range(0,15)
y = [1/(n+1) for n in x]
plt.plot(x,y)

plt.title('$ y = 1/(x+1)$')
plt.xlabel('$x$')
plt.ylabel('$1/x$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0);

plt.tight_layout()
```

```python
x = list(range(0, 25))

# generate y for each fucntion
y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

plt.plot(x, y1, color ='brown', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x, y2, color ='blue', label='$x^{3}$', alpha=.8)
plt.plot(x, y3, color = 'red', label='$2^{x}$', alpha=.8)
plt.plot(x, y4, color = 'green', label='$1/(x+1)$', alpha=.8)


plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('y')
plt.xlabel('x')
plt.title('My Functions')
plt.legend()

plt.show()
```
