```python
import numpy as np
```

```python
# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
# Perform an outer join specifying the left and right DataFrame keys.

users.merge(roles, left_on='role_id', right_on='id', how='outer', indicator=True)
```

```python
(users.merge(roles, 
            left_on='role_id', 
            right_on='id', 
            how='outer')
    .drop(columns='role_id')
    .rename(columns={'id_x': 'id', 
                     'name_x': 'employee',
                     'id_y': 'role_id',
                     'name_y': 'role'}
            )
)
```

```python
#2.2
users.merge(roles, 
            how="right", 
            left_on="role_id", 
            right_on="id")
```

```python
right_join = pd.merge(users,
                     roles,
                     left_on='role_id',
                     right_on='id',
                      how='right',
                     indicator=True)
right_join
```

```python
#2.3
users.merge(roles, 
            how="outer", 
            left_on="role_id", 
            right_on="id")
```

```python
pd.merge(users,
        roles,
        left_on='role_id',
        right_on='id',
        how='outer',
        indicator=True)
```

```python
#2.4
users.drop(columns="role_id").merge(roles, how="outer", on="id")
```

```python
#bad idea to do this

pd.merge(users,
        roles,
        left_on='id',
        right_on='id',
        how='outer',
        indicator=True)
```

```python
#2.12 & 13
df['mileage_difference'] = df.hwy_mpg - df.city_mpg

df['average_mileage'] = (df.city_mpg + df.hwy_mpg) / 2

df.head()
```

```python
df["average_mileage"] = df[["city_mpg", "hwy_mpg"]].mean(axis=1)
df.head(2)
```

```python
#2.14

df["is_automatic"] = df.trans.str.contains("auto")
df.head(2)
```

```python
df.trans = np.where(df.trans.str.startswith("auto"), "automatic", "manual")
df.head()
```

```python
#2.15

manufacturer_avg_mpg = df.groupby("manufacturer").average_mileage.mean()

manufacturer_avg_mpg.idxmax()
```
